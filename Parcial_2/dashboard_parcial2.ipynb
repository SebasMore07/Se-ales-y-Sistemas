```python
%%writefile 1_üìà_An√°lisis_del_Espectro_de_Fourier.py

import streamlit as st

import numpy as np
import matplotlib.pyplot as plt


# Configuraci√≥n de la p√°gina en Streamlit
st.set_page_config(page_title="An√°lisis del Espectro de Fourier", page_icon="üìà")

st.markdown("# An√°lisis del Espectro de Fourier")
st.sidebar.header("Configuraci√≥n de Par√°metros")


# Secci√≥n Te√≥rica
st.markdown("## 1. Expresi√≥n del Espectro de Fourier")
st.write("El espectro de Fourier para la se√±al se obtiene utilizando identidades trigonom√©tricas y la serie de Fourier.")

st.markdown("### 1.1. Simplificaci√≥n de la Se√±al")
st.latex(r"x(t) = |A\cos(2\pi F_o t)|^2")
st.write("Usamos la identidad trigonom√©trica:")
st.latex(r"\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}")
st.write("Para obtener:")
st.latex(r"x(t) = \frac{A^2}{2} + \frac{A^2}{2} \cos(4\pi F_o t)")

st.markdown("### 1.2. Coeficientes de Fourier")
st.write("La serie de Fourier en forma exponencial es:")
st.latex(r"x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j 2\pi n F_o t}")
st.write("Los coeficientes son:")
st.latex(r"c_0 = \frac{A^2}{2}, \quad c_2 = c_{-2} = \frac{A^2}{4}, \quad c_n = 0 \text{ para otros } n.")
st.write("La representaci√≥n en forma trigonom√©trica es:")
st.latex(r"x(t) = \frac{A^2}{2} + \frac{A^2}{2} \cos(4\pi F_o t)")

st.markdown("### 1.3. Espectro de Fourier")
st.write("El espectro tiene componentes en:")
st.latex(r"f = -2F_o, 0, 2F_o")
st.write("Con magnitudes:")
st.latex(r"\frac{A^2}{4}, \frac{A^2}{2}, \frac{A^2}{4}")
st.write("Y fase 0 en todos los casos.")


#mostrar la se√±al
Fo = 60
T = 1/Fo
Fs = 100*Fo # frecuencia de muestreo definida
tv = np.arange(-T/2,T/2,1/Fs) #vector de tiempo generado
Nm = len(tv) #numero de muestras requeridas
A=1 #Constante definida
x = (A*np.cos(2*np.pi*Fo*tv))**2

#graficar se√±al original
fig, ax = plt.subplots()
ax.plot(tv, x, 'r', linewidth=4)
ax.grid()
ax.set_xlabel("t [s]", fontsize=14)
ax.set_ylabel("x(t)", fontsize=14)
ax.set_title("Gr√°fica de x(t)")

# Mostrar la gr√°fica en Streamlit
st.pyplot(fig)

N = 50 #numero armonicos
wo = 2*np.pi/T # frecuencia fundamental
# definir bases
phin = np.zeros((Nm,2*N+1),dtype=np.complex_)#crear matriz para guardar bases
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)#base de fourier en el intervalo¬†de¬†inter√©s


#calcular espectro -> cambiar seg√∫n la se√±al estudiada
cn = np.zeros(2*N+1,dtype=np.complex_)
nv = np.linspace(-N, N, 2*N+1) #vector num arm√≥nicos
Cn = 0*nv
Cn[N] = (A**2)/2 #Nivel DC
Cn[N+2] = (A**2)/4
Cn[N-2]=(A**2)/4

#graficar espectro
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Graficar la parte real
axes[0, 0].stem(nv, np.real(Cn), 'r')
axes[0, 0].set_xlabel(r'$n$', fontsize=14)
axes[0, 0].set_ylabel(r'$Re\{C_n\}$', fontsize=14)
axes[0, 0].grid()
axes[0, 0].set_title("Parte Real")

# Graficar la parte imaginaria
axes[0, 1].stem(nv, np.imag(Cn), 'r')
axes[0, 1].set_xlabel(r'$n$', fontsize=14)
axes[0, 1].set_ylabel(r'$Im\{C_n\}$', fontsize=14)
axes[0, 1].grid()
axes[0, 1].set_title("Parte Imaginaria")

# Graficar la magnitud del espectro
axes[1, 0].stem(nv, abs(Cn), 'r')
axes[1, 0].set_xlabel(r'$n$', fontsize=14)
axes[1, 0].set_ylabel(r'$|C_n|$', fontsize=14)
axes[1, 0].grid()
axes[1, 0].set_title("Magnitud del Espectro")

# Graficar la fase del espectro
phase = np.zeros_like(Cn)
for i in range(len(Cn)):
    if np.imag(Cn[i]) == 0:
        phase[i] = 0  # Fase cero cuando la parte imaginaria es cero
    else:
        phase[i] = np.angle(Cn[i])

axes[1, 1].stem(nv, phase, 'r')
axes[1, 1].set_xlabel(r'$n$', fontsize=14)
axes[1, 1].set_ylabel(r'$\langle C_n$', fontsize=14)
axes[1, 1].grid()
axes[1, 1].set_title("Fase del Espectro")

# Ajustar el dise√±o y mostrar en Streamlit
fig.tight_layout()
st.pyplot(fig)



fig, axes = plt.subplots(2, 1, figsize=(8, 8))

# Graficar la magnitud en dB
axes[0].stem(nv, 20 * np.log10(abs(Cn)), 'r')
axes[0].set_xlabel(r'$n$', fontsize=14)
axes[0].set_ylabel(r'$|C_n|$ [dB]', fontsize=14)
axes[0].grid()
axes[0].set_title("Magnitud del Espectro (dB)")

# Graficar la fase ajustada manualmente
phase = np.zeros_like(Cn, dtype=float)
for i in range(len(Cn)):
    if np.imag(Cn[i]) == 0:
        phase[i] = 0  # Fase cero cuando la parte imaginaria es cero
    else:
        phase[i] = np.angle(Cn[i], deg=True)  # Calcular la fase en grados

axes[1].stem(nv, phase, 'r')
axes[1].set_xlabel(r'$n$', fontsize=14)
axes[1].set_ylabel(r'$Fase$ [¬∞]', fontsize=14)
axes[1].grid()
axes[1].set_title("Fase del Espectro")

# Ajustar el dise√±o y mostrar en Streamlit
fig.tight_layout()
st.pyplot(fig)



# Definir par√°metros iniciales
Px = 0.375  # Estimaci√≥n de la potencia
N_max = len(Cn) // 2  # M√°ximo valor de Na basado en Cn

# Crear slider en Streamlit
Na = st.slider("Selecciona Na:", min_value=0, max_value=N_max, value=1, step=1)

# Calcular √≠ndice y error
ind = np.arange(N - Na, N + Na + 1)
er = 1 - np.sum(abs(Cn[ind]) ** 2) / Px

# Se√±al reconstruida
xe = phin[:, ind].dot(Cn[ind])  # Producto entre vectores

# Graficar en Streamlit
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(tv, xe, color='b', marker='o', markersize=5, label='$x_e(t)$')  # Se√±al estimada
ax.plot(tv, x, color='r', label='x(t)')  # Se√±al original
ax.set_title(f'$E_r$ = {100 * er:.2f}%', fontsize=16)
ax.set_xlabel('t [s]')
ax.set_ylabel('x(t)')
ax.grid()
ax.legend()

# Mostrar gr√°fica en Streamlit
st.pyplot(fig)



st.markdown("## 4. Conclusi√≥n")
st.write("""
El an√°lisis del espectro de Fourier muestra que la se√±al tiene componentes en $( f = -2F_o, 0, 2F_o )$, con magnitudes $frac{A^2}{4}$, $frac{A^2}{2}$, $frac{A^2}{4}$. La reconstrucci√≥n de la se√±al con una serie finita de t√©rminos demuestra que con pocas componentes se puede recuperar la forma original.
""")


```

##TERCER PUNTO

```python
%%writefile 3_‚ö°_Total_Harmonic_Distortion_THD.py

import streamlit as st
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq


# Configuraci√≥n de la p√°gina
st.set_page_config(page_title="Distorsi√≥n Total de Arm√≥nicos (THD)", page_icon="‚ö°")

# T√≠tulo y descripci√≥n
st.title("Distorsi√≥n Total de Arm√≥nicos (THD) y Factor de Potencia")
st.sidebar.header("Configuraci√≥n del Circuito")


st.header("1. Distorsi√≥n Total de Arm√≥nicos (THD)")
st.write(
    """
    La **Distorsi√≥n Total de Arm√≥nicos (THD)** mide la distorsi√≥n en una se√±al peri√≥dica debido a la presencia de arm√≥nicos.
    Se calcula como:
    """
)
st.latex(r"""
THD = \frac{\sqrt{\sum_{n=2}^{\infty} V_n^2}}{V_1} \times 100\%
""")
st.write(
    "Donde $V_1$ es la magnitud de la frecuencia fundamental y $V_n$ las magnitudes de los arm√≥nicos."
)

st.header("2. C√°lculo del THD desde la FFT")
st.write(
    """
    La Transformada R√°pida de Fourier (FFT) permite calcular el THD descomponiendo la se√±al en sus componentes de frecuencia.
    """
)
st.latex(r"""
THD = \frac{\sqrt{V_2^2 + V_3^2 + \dots + V_n^2}}{V_1} \times 100\%
""")

st.header("3. Factor de Potencia y Distorsi√≥n")
st.write("El factor de potencia (FP) mide la eficiencia en el uso de la energ√≠a:")

st.latex(r"""
PF_{THD} = \sqrt{\frac{1}{1 + (THD)^2}}
""")
st.header("4. Simulaci√≥n de THD y Factor de Potencia")

# Selecci√≥n del tipo de carga
carga_tipo = st.selectbox("Seleccione el tipo de carga:", ["Resistiva", "RC en serie"])

Fo = 60  # Frecuencia de alimentaci√≥n
Fs = 30 * Fo  # Frecuencia de muestreo
To = 1 / Fo  # Per√≠odo fundamental
Ts = 1 / Fs  # Per√≠odo de muestreo
t = np.arange(0, 5 * To, Ts)  # Simulaci√≥n de 5 per√≠odos
A = 120  # Amplitud de la entrada

# Se√±al de entrada
in_o = A * np.sin(2 * np.pi * Fo * t)
rec_c = sig.square(2 * np.pi * Fo * t)  # Tren de pulsos
in_ = in_o * rec_c  # Entrada rectificada completa

if carga_tipo == "Resistiva":
    R_v = st.number_input("Ingrese el valor de R:", min_value=0.1, value=10.0)

    num = np.array([1])
    den = np.array([1])
    G_n = sig.TransferFunction(num, den)
    _, out, _ = sig.lsim(G_n, U=in_, T=t)

elif carga_tipo == "RC en serie":
    R_v = st.number_input("Ingrese el valor de R:", min_value=0.1, value=10.0)
    C_v = st.number_input("Ingrese la capacitancia (F):", min_value=1e-9, value=1e-6, step=1e-6, format="%.6f")

    num = np.array([1])
    den = np.array([R_v * C_v, 1])
    G_n = sig.TransferFunction(num, den)
    _, out, _ = sig.lsim(G_n, U=in_, T=t)
    #out = G_n.output(in_, T=t)[1]
# Graficar la se√±al de entrada y salida
fig, ax = plt.subplots()
ax.plot(t, in_, label='Entrada')
ax.plot(t, out, label='Salida', linestyle='dashed')
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Amplitud')
ax.legend()
st.pyplot(fig)

# C√°lculo del espectro
vfre = np.fft.rfftfreq(len(out), 1 / Fs)
Xf = np.fft.rfft(out)

fig, ax = plt.subplots()
ax.stem(vfre, abs(Xf) / len(out), basefmt=" ")
ax.set_xlabel('Frecuencia [Hz]')
ax.set_ylabel('|X(f)|')
ax.set_title('Espectro de Frecuencia')
ax.grid()
st.pyplot(fig)

# C√°lculo del THD
def calcular_thd(signal, Fs):
    Xf = np.fft.rfft(signal)
    magnitudes = np.abs(Xf)
    if len(magnitudes) < 2:
        return 0
    fundamental_idx = np.argmax(magnitudes)
    fundamental = magnitudes[fundamental_idx]
    if fundamental == 0:
        return 0
    thd = np.sqrt(np.sum(magnitudes[1:] ** 2)) / fundamental
    return thd

thd_value = calcular_thd(out, Fs)
st.write(f"**THD del circuito {carga_tipo}:** {thd_value * 100:.2f}%")

pf = np.sqrt(1 / (1 + thd_value ** 2))
st.write(f"**El factor de potencia de la se√±al con carga {carga_tipo}: **{pf:.4f}")
```

```python
%%writefile 2_üì°üéµ_Modulaci√≥n_en_Amplitud_AM.py


import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import yt_dlp
import soundfile as sf
from io import BytesIO
import tempfile
import os
from scipy.io.wavfile import write
import io


# Configuraci√≥n inicial
st.title("Modulaci√≥n AM")

# Par√°metros
Fc = 100  # Frecuencia de la portadora (Hz)
Fm = 10   # Frecuencia del mensaje (Hz)
Am = 1    # Amplitud del mensaje

# Definici√≥n del tiempo
ti = 0
tf = 5 / Fm  # N√∫mero de per√≠odos del mensaje a graficar
Fs = 5 * Fc  # Frecuencia de muestreo
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo

# Generaci√≥n de se√±ales
m = Am * np.cos(2 * np.pi * Fm * t)  # Se√±al del mensaje
Im = 0.8  # √çndice de modulaci√≥n
Ac = max(abs(m)) / Im  # Amplitud de la portadora seg√∫n el √≠ndice de modulaci√≥n
c = Ac * np.sin(2 * np.pi * Fc * t)  # Se√±al portadora

y = (1 + m / Ac) * c  # Se√±al modulada en AM

# Gr√°fico 1: Se√±al de la portadora y mensaje
st.subheader("Se√±al de la portadora y mensaje")
fig1, ax1 = plt.subplots()
ax1.plot(t, c, label='Portadora')
ax1.plot(t, m, label='Mensaje')
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Amplitud")
ax1.set_title("Portadora y mensaje en el tiempo")
ax1.legend()
st.pyplot(fig1)

# Explicaci√≥n del gr√°fico 1
st.write("Este gr√°fico muestra la se√±al de la portadora (onda senoidal de alta frecuencia) y la se√±al del mensaje (onda cosenoidal de menor frecuencia). La portadora sirve como la base sobre la cual se modula la informaci√≥n del mensaje.")

# Gr√°fico 2: Se√±al modulada en AM
st.subheader("Se√±al modulada en AM")
fig2, ax2 = plt.subplots()
ax2.plot(t, y, label='Se√±al AM')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Amplitud")
ax2.set_title("Se√±al modulada en amplitud")
ax2.legend()
st.pyplot(fig2)

# Explicaci√≥n del gr√°fico 2
st.write("En este gr√°fico se muestra la se√±al resultante despu√©s de la modulaci√≥n en amplitud (AM). La envolvente de la se√±al modulada sigue la forma de la se√±al del mensaje, lo que permite recuperar la informaci√≥n original en el receptor.")

# Transformada de Fourier
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)

vfre = np.fft.fftfreq(len(c), 1 / Fs)

# Gr√°fico 3: Espectro de frecuencia
st.subheader("Espectro de frecuencia de las se√±ales")
fig3, ax3 = plt.subplots()
ax3.plot(vfre, abs(Xfy), label='Modulado')
ax3.plot(vfre, abs(Xfm), label='Mensaje')
ax3.plot(vfre, abs(Xfc), label='Portadora')
ax3.set_xlabel("Frecuencia (Hz)")
ax3.set_ylabel("|X(f)|")
ax3.set_title("An√°lisis espectral de las se√±ales")
ax3.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax3.legend()
st.pyplot(fig3)

# Explicaci√≥n del gr√°fico 3
st.write("Este gr√°fico muestra el espectro de frecuencia de las se√±ales. Se puede observar la se√±al del mensaje en bajas frecuencias, la portadora en su frecuencia correspondiente y la se√±al modulada con sus componentes espectrales en torno a la portadora.")

#Canci√≥n

# Configuraci√≥n de la app
st.title("Descarga y reproducci√≥n de audio desde YouTube")

# Informaci√≥n de la canci√≥n
nombre_cancion = "Queen - Killer Queen"
st.write(f"**Canci√≥n:** {nombre_cancion}")

# Descarga del audio desde YouTube
def descargar_audio(url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'extractaudio': True,
        'audioformat': 'mp3',
        'outtmpl': 'tema.mp3'
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

def convertir_a_wav():
    os.system("ffmpeg -y -i tema.mp3 output.wav")

# URL del video de YouTube
link = "https://www.youtube.com/watch?v=9K-Jmg93Xj8"
st.write("Descargando audio...")
descargar_audio(link)
st.write("Convirtiendo a WAV...")
convertir_a_wav()

# Cargar el audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

# Mostrar la frecuencia de muestreo
st.write(f"**Frecuencia de muestreo:** {fs} Hz")

# Selecci√≥n del fragmento de audio (segundos 20 a 25)
inicio, fin = int(fs * 20), int(fs * 25)
segmento_audio = x[inicio:fin]

# Guardar el segmento en un archivo temporal
with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as temp_wav:
    sf.write(temp_wav.name, segmento_audio, fs)
    temp_wav_path = temp_wav.name

# Reproducir el audio en Streamlit
st.audio(temp_wav_path, format="audio/wav")

st.write("Este audio corresponde al segmento entre los segundos 20 y 25 de la canci√≥n.")

# Configuraci√≥n de la app
st.title("Procesamiento de Audio y Modulaci√≥n AM")

# Cargar el audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

# Seleccionar el canal izquierdo y fragmento de audio (segundos 20 a 25)
ti, tf = 20, 25
xpro = x.copy()
xs = xpro[int(ti * fs):int(tf * fs), 0]

# Vector de tiempo
tt = np.arange(ti, tf, 1 / fs)

# Gr√°fico 1: Se√±al original (canal izquierdo)
st.subheader("Se√±al Original del Canal Izquierdo")
fig1, ax1 = plt.subplots()
ax1.plot(tt, xs, label="Canal izquierdo")
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Amplitud")
ax1.set_title("Se√±al de audio entre 20s y 25s")
ax1.legend()
st.pyplot(fig1)

# Explicaci√≥n del gr√°fico 1
st.write("Este gr√°fico muestra la se√±al original del canal izquierdo en el intervalo de tiempo entre 20 y 25 segundos.")

# Entrada del usuario para el √≠ndice de modulaci√≥n
Im = st.number_input("Ingrese el √≠ndice de modulaci√≥n:", min_value=0.1, max_value=2.0, value=0.8, step=0.1)

# Par√°metros de modulaci√≥n AM
Fc = 1000  # Frecuencia de la portadora (Hz)
Fs = 48000  # Frecuencia de muestreo
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo

# Generar la portadora
Ac = max(abs(xs)) / Im  # Amplitud de la portadora seg√∫n √≠ndice de modulaci√≥n
c = Ac * np.sin(2 * np.pi * Fc * t)  # Se√±al portadora

# Gr√°fico 2: Se√±al portadora y mensaje
st.subheader("Se√±al Portadora y Mensaje")
fig2, ax2 = plt.subplots()
ax2.plot(t, c, label='Portadora')
ax2.plot(t, xs, label='Mensaje')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Amplitud")
ax2.set_title("Comparaci√≥n entre portadora y mensaje")
ax2.legend()
st.pyplot(fig2)

# Explicaci√≥n del gr√°fico 2
st.write("En este gr√°fico se observa la se√±al portadora (onda senoidal de alta frecuencia) y la se√±al del mensaje original (audio extra√≠do). La portadora se usa para modular la se√±al de audio.")

# Se√±al modulada en AM
y = (1 + xs / Ac) * c

# Gr√°fico 3: Se√±al modulada AM
st.subheader("Se√±al Modulada en AM")
fig3, ax3 = plt.subplots()
ax3.plot(t, y, label='Se√±al Modulada AM')
ax3.set_xlabel("Tiempo (s)")
ax3.set_ylabel("Amplitud")
ax3.set_title("Modulaci√≥n AM de la Se√±al de Audio")
ax3.legend()
st.pyplot(fig3)

st.write("Este gr√°fico muestra la se√±al modulada en amplitud (AM). La forma de la se√±al refleja c√≥mo la envolvente sigue la se√±al del mensaje, lo que permite transmitir la informaci√≥n del audio.")

# Transformada de Fourier
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# Gr√°fico 4: Espectro de la portadora
st.subheader("Espectro de Frecuencia de la Portadora")
fig4, ax4 = plt.subplots()
ax4.plot(vfre, abs(Xfc), label='Portadora')
ax4.set_xlabel("Frecuencia (Hz)")
ax4.set_ylabel("|X(f)|")
ax4.set_title("Espectro de Frecuencia de la Se√±al Portadora")
ax4.legend()
st.pyplot(fig4)
st.write("La se√±al portadora es una onda sinusoidal pura de frecuencia Fc = 1000 Hz. Su espectro de frecuencia muestra un √∫nico pico en esta frecuencia, ya que una onda sinusoidal tiene toda su energ√≠a concentrada en una sola frecuencia")

# Gr√°fico 5: Espectro de la Se√±al Modulada AM
st.subheader("Espectro de Frecuencia de la Se√±al Modulada")
fig5, (ax5a, ax5b) = plt.subplots(2, 1, figsize=(8, 8), constrained_layout=True)  # Aumentando el espacio
ax5a.plot(vfre, abs(Xfy), label='Modulado', color='r')
ax5a.set_xlabel("Frecuencia (Hz)")
ax5a.set_ylabel("|X(f)|")
ax5a.set_title("Espectro de la Se√±al Modulada AM")
ax5a.legend()
ax5a.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])

ax5b.plot(vfre, abs(Xfm), label='Mensaje')
ax5b.set_xlabel("Frecuencia (Hz)")
ax5b.set_ylabel("|X(f)|")
ax5b.set_title("Espectro de la Se√±al del Mensaje")
ax5b.legend()
ax5b.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])

st.pyplot(fig5)

st.write("El primer gr√°fico del espectro muestra la se√±al modulada en AM, donde se pueden ver las bandas laterales. El segundo gr√°fico muestra el espectro del mensaje original antes de la modulaci√≥n.")

# Funci√≥n para convertir el audio en un buffer
def convert_audio(data, rate):
    buffer = io.BytesIO()
    write(buffer, rate, data.astype(np.float32))
    buffer.seek(0)  # Reiniciar el puntero para asegurar la reproducci√≥n
    return buffer

# Convertir los audios antes de mostrarlos para reducir la demora
buffer_c = convert_audio(c, fs)
buffer_xs = convert_audio(xs, fs)
buffer_y = convert_audio(y, fs)

# Crear pesta√±as para seleccionar el audio
tab1, tab2, tab3 = st.tabs(["Portadora", "Mensaje", "Modulada"])

with tab1:
    st.subheader("Audio de la portadora")
    st.audio(buffer_c, format='audio/wav')

with tab2:
    st.subheader("Audio del mensaje original")
    st.audio(buffer_xs, format='audio/wav')

with tab3:
    st.subheader("Audio de la se√±al modulada")
    st.audio(buffer_y, format='audio/wav')
```

```python
%%writefile 4_üéôÔ∏è_Reconocimiento_de_voz.py

from google.colab import files  # Para la carga de archivos
import streamlit as st
import gdown
import os
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
import librosa
import scipy.fftpack as fft
import pickle
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import pandas as pd
from scipy import fft
from pydub import AudioSegment


# Configuraci√≥n de la p√°gina
st.set_page_config(page_title="Reconocimiento de Voz", page_icon="üéôÔ∏è")

# T√≠tulo y descripci√≥n
st.markdown("# Reconocimiento de Voz üéôÔ∏è")
st.sidebar.header("Exploraci√≥n del Reconocimiento de Voz")


# ID del archivo en Google Drive (Aseg√∫rate de que el archivo sea p√∫blico)
FILE_ID = "1TZBmNzbGLTOS2kXAXngV-5Yf6MrzMrJb"
FILENAME = "audioClaveSe√±al.wav"

# Enlace de descarga generado
DOWNLOAD_URL = f"https://drive.google.com/uc?export=download&id={FILE_ID}"

st.title("üîä Descarga y Reproducci√≥n de Audio con palabra clave")
st.write("Este m√≥dulo descarga un archivo de audio desde Google Drive y permite escucharlo directamente.")

# Descargar el archivo si no existe
if not os.path.exists(FILENAME):
    with st.spinner("Descargando el archivo de audio..."):
        gdown.download(DOWNLOAD_URL, FILENAME, quiet=False)
    st.success("Descarga completada ‚úÖ")

# Reproducir el archivo de audio
if os.path.exists(FILENAME):
    st.audio(FILENAME, format="audio/wav")
    st.success("Reproduciendo el audio... üéß")
else:
    st.error("No se pudo descargar el archivo. Verifica el ID de Google Drive.")

#------------------------------------------------------------------------------------------

# Ruta del archivo de audio predefinido
file_path = "audioClaveSe√±al.wav"

# Leer el archivo de audio
try:
    audio_data, sample_rate = sf.read(file_path)

    # Mostrar informaci√≥n b√°sica del audio
    st.write(f"üìå **Frecuencia de muestreo:** {sample_rate} Hz")
    st.write(f"üìå **N√∫mero de muestras:** {len(audio_data)}")

    # Si el audio es est√©reo, convertirlo a mono
    if len(audio_data.shape) == 2:
        audio_data = audio_data.mean(axis=1)
        st.write("üîÑ Audio convertido a **mono**.")

    # Duraci√≥n del audio en segundos
    duracion = len(audio_data) / sample_rate
    st.write(f"‚è±Ô∏è **Duraci√≥n del audio:** {duracion:.2f} segundos")

    # Graficar la se√±al de audio
    t = np.linspace(0, duracion, len(audio_data))  # Vector de tiempo
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.plot(t, audio_data, color="blue")
    ax.set_title("Se√±al de Audio")
    ax.set_xlabel("Tiempo [s]")
    ax.set_ylabel("Amplitud")
    ax.grid()

    # Mostrar la gr√°fica en Streamlit
    st.pyplot(fig)

    # Reproducir el audio en la aplicaci√≥n
    #st.audio(file_path, format="audio/wav")

except FileNotFoundError:
    st.error("‚ùå El archivo de audio no fue encontrado. Verifica la ruta y aseg√∫rate de que existe.")


#------------------------------------------------------------------------------------

#DIVIDIMOS EL AUDIO ORIGINAL EN SEGMENTOS POR CADA PALABRA CLAVE

# Definir la funci√≥n para segmentar el audio
def segment_audio_by_energy(audio_path, output_folder, frame_length=1024, hop_length=256, padding=0.05, merge_gap=0.3):
    y, sr = librosa.load(audio_path, sr=16000, mono=True)
    energy = np.array([np.sum(np.square(y[i : i + frame_length])) for i in range(0, len(y), hop_length)])
    energy = energy / np.max(energy)
    energy_threshold = np.percentile(energy, 55)
    voiced_frames = energy > energy_threshold
    change_points = np.diff(voiced_frames.astype(int))
    start_indices = np.where(change_points == 1)[0] * hop_length
    end_indices = np.where(change_points == -1)[0] * hop_length

    if len(end_indices) == 0 or (len(start_indices) > 0 and start_indices[0] > end_indices[0]):
        end_indices = np.append(end_indices, len(y))
    if len(start_indices) == 0 or (len(end_indices) > 0 and end_indices[-1] < start_indices[-1]):
        start_indices = np.insert(start_indices, 0, 0)

    min_duration = 0.1 * sr
    valid_segments = [(start, end) for start, end in zip(start_indices, end_indices) if end - start > min_duration]

    merged_segments = []
    prev_start, prev_end = valid_segments[0]
    for start, end in valid_segments[1:]:
        if start - prev_end < merge_gap * sr:
            prev_end = end
        else:
            merged_segments.append((prev_start, prev_end))
            prev_start, prev_end = start, end
    merged_segments.append((prev_start, prev_end))

    padding_samples = int(padding * sr)
    merged_segments = [(max(0, start - padding_samples), min(len(y), end + padding_samples)) for start, end in merged_segments]

    os.makedirs(output_folder, exist_ok=True)

    segment_links = []
    for i, (start, end) in enumerate(merged_segments):
        segment = y[start:end]
        output_file = os.path.join(output_folder, f'segment_{i+1}.wav')
        sf.write(output_file, segment, sr)
        segment_links.append((f"audioClave{i+1}", output_file, 1))

    return segment_links

# Uso del c√≥digo
input_audio = "audioClaveSe√±al.wav"
output_directory = "segmented_audios"
segments_data = segment_audio_by_energy(input_audio, output_directory)


#-------------------------------------------------------------------------------


#AUDIO ERROR

FILEID="1hNUfuAtLJr4KOJFgrpVtIe7xV8GRBl4X"
FILENAME2="audioError.wav"

# Enlace de descarga generado
DOWNLOAD_URL = f"https://drive.google.com/uc?export=download&id={FILEID}"

st.title("üîä Descarga y Reproducci√≥n de Audio con palabras error")
st.write("Este m√≥dulo descarga un archivo de audio desde Google Drive y permite escucharlo directamente.")

# Descargar el archivo si no existe
if not os.path.exists(FILENAME2):
    with st.spinner("Descargando el archivo de audio..."):
        gdown.download(DOWNLOAD_URL, FILENAME2, quiet=False)
    st.success("Descarga completada ‚úÖ")

# Reproducir el archivo de audio
if os.path.exists(FILENAME2):
    st.audio(FILENAME2, format="audio/wav")
    st.success("Reproduciendo el audio... üéß")
else:
    st.error("No se pudo descargar el archivo. Verifica el ID de Google Drive.")


#SEGMENTAR AUDIO ERROR


input_audio = "audioError.wav"
output_directory = "segmented_audiosError"
segment=segment_audio_by_energy(input_audio, output_directory)


#--------------------------------------------------------------------------------------------

# Funci√≥n para calcular el espectro de Fourier
def compute_spectrum(signal, sr, n_fft=1024):
    spectrum = np.abs(fft.fft(signal, n=n_fft))  # Magnitud del espectro
    min_val = np.min(spectrum)
    if min_val == 0:
        min_val = 1e-10  # Evitar divisi√≥n por cero
    spectrum = spectrum / min_val  # Normalizar por el valor m√≠nimo
    return spectrum[:n_fft // 2]  # Usamos solo la mitad (simetr√≠a de la FFT)

# Funci√≥n para procesar una carpeta de audios y crear la base de datos
def create_database(folder_path, sr=22050):
    database = []
    filenames = []

    if not os.path.exists(folder_path):
        st.error(f"La carpeta {folder_path} no existe.")
        return np.array([]), []

    files = [f for f in os.listdir(folder_path) if f.endswith(".wav")]
    if not files:
        st.warning(f"No se encontraron archivos .wav en {folder_path}")
        return np.array([]), []

    for filename in files:
        file_path = os.path.join(folder_path, filename)
        signal, _ = librosa.load(file_path, sr=sr)  # Cargar el audio
        spectrum = compute_spectrum(signal, sr)  # Calcular el espectro
        database.append(spectrum)  # Agregar a la base de datos
        filenames.append(filename)  # Guardar el nombre del archivo

    return np.array(database), filenames

# Rutas de las carpetas
correct_folder = "segmented_audios"  # Carpeta con la palabra correcta
error_folder = "segmented_audiosError"  # Carpeta con palabras incorrectas

# Cargar bases de datos
st.write("üîÑ Cargando bases de datos de audios...")
correct_database, correct_filenames = create_database(correct_folder)
error_database, error_filenames = create_database(error_folder)

# Verificar si hay datos cargados
if correct_database.size > 0 and error_database.size > 0:
    # Combinar bases de datos
    X = np.vstack((correct_database, error_database))  # Datos de espectros
    labels = np.array([1] * len(correct_database) + [0] * len(error_database))  # Etiquetas
    filenames = correct_filenames + error_filenames  # Nombres de archivos

    st.success("‚úÖ Bases de datos cargadas correctamente")
    st.write(f"üìÇ Total de archivos procesados: {len(filenames)}")

    # Mostrar tabla con los nombres de archivos y sus etiquetas
    df = pd.DataFrame({"Archivo": filenames, "Etiqueta (1 = Correcta, 0 = Incorrecta)": labels})
    st.dataframe(df)

    # Seleccionar un archivo para visualizar su espectro
    selected_file = st.selectbox("üîç Selecciona un archivo para ver su espectro:", filenames)

    if selected_file:
        index = filenames.index(selected_file)
        spectrum = X[index]

        # Graficar el espectro de Fourier
        fig, ax = plt.subplots()
        ax.plot(spectrum)
        ax.set_title(f"Espectro de Fourier - {selected_file}")
        ax.set_xlabel("Frecuencia")
        ax.set_ylabel("Magnitud")
        st.pyplot(fig)

else:
    st.error("‚ùå No se pudieron cargar los audios. Verifica las carpetas y los archivos.")
#-------------------------------------------------------------------------------------------------------

# Reducci√≥n a 2D usando t-SNE o PCA
fmax = 7000  # Considerar solo frecuencias hasta fmax Hz

if 'X' in locals() and 'labels' in locals():
    red_ = TSNE(perplexity=20, n_components=2, random_state=123, learning_rate='auto', init='pca')
    # Alternativa: PCA
    # red_ = PCA(n_components=2, random_state=123)

    X_2D = red_.fit_transform(X[:, :fmax])  # Reducir a 2D

    # Graficar separabilidad 2D
    fig, ax = plt.subplots(figsize=(10, 8))
    scatter = ax.scatter(X_2D[:, 0], X_2D[:, 1], c=labels, cmap='viridis', s=50)
    plt.colorbar(scatter, label="Etiquetas (1: Correcta, 0: Incorrecta)")
    ax.set_title("Separabilidad de los datos en 2D")
    ax.set_xlabel("Componente 1")
    ax.set_ylabel("Componente 2")
    st.pyplot(fig)

    # Graficar con nombres de archivos
    fig, ax = plt.subplots(figsize=(12, 8))
    colors = ["blue", "yellow"]  # Colores para las etiquetas
    scatter = ax.scatter(X_2D[:, 0], X_2D[:, 1], c=labels, cmap='viridis', s=50)

    for i, tex in enumerate(filenames):
        ax.text(X_2D[i, 0] * 1.025, X_2D[i, 1] * 1.025, tex[:-4] + f"_{i}", fontsize=8, color=colors[labels[i]])

    ax.set_title("Separabilidad de los datos en 2D con nombres de archivos")
    ax.set_xlabel("Componente 1")
    ax.set_ylabel("Componente 2")
    st.pyplot(fig)

    # Guardar las bases de datos en archivos
    with open("correct_database.pkl", "wb") as f:
        pickle.dump(correct_database, f)

    with open("error_database.pkl", "wb") as f:
        pickle.dump(error_database, f)

    st.success("‚úÖ Bases de datos creadas y guardadas correctamente.")
    st.write(f"üìÇ Tama√±o de la base de datos correcta: {len(correct_database)}")
    st.write(f"üìÇ Tama√±o de la base de datos de errores: {len(error_database)}")
else:
    st.error("‚ùå No se encontraron datos para reducir a 2D. Verifica la carga de la base de datos.")


#--------------------------------------------------------------------------------------------------------

# Funci√≥n para convertir un archivo de audio a WAV
def convert_to_wav(input_file, output_file):
    audio = AudioSegment.from_file(input_file)
    audio.export(output_file, format="wav")
    return output_file

# Funci√≥n para calcular el espectro de Fourier
def compute_spectrum(signal, sr, n_fft=1024):
    spectrum = np.abs(fft.fft(signal, n=n_fft))  # Magnitud del espectro
    spectrum = spectrum[:n_fft // 2]  # Usamos solo la mitad (simetr√≠a de la FFT)
    spectrum = spectrum / np.max(spectrum)  # Normalizar el espectro
    return spectrum

# Funci√≥n para calcular el √≠ndice del valor m√≠nimo en el espectro
def compute_min_spectrum_index(signal, sr, n_fft=1024):
    spectrum = compute_spectrum(signal, sr, n_fft)
    min_index = np.argmin(spectrum)  # √çndice del valor m√≠nimo en el espectro
    return min_index

# Funci√≥n para segmentar audio por energ√≠a
def segment_audio_by_energy(y, sr, frame_length=1024, hop_length=256, padding=0.05, merge_gap=0.3):
    energy = np.array([
        np.sum(np.square(y[i : i + frame_length]))
        for i in range(0, len(y), hop_length)
    ])
    energy = energy / np.max(energy)
    energy_threshold = np.percentile(energy, 55)
    voiced_frames = energy > energy_threshold
    change_points = np.diff(voiced_frames.astype(int))
    start_indices = np.where(change_points == 1)[0] * hop_length
    end_indices = np.where(change_points == -1)[0] * hop_length
    if len(end_indices) == 0 or (len(start_indices) > 0 and start_indices[0] > end_indices[0]):
        end_indices = np.append(end_indices, len(y))
    if len(start_indices) == 0 or (len(end_indices) > 0 and end_indices[-1] < start_indices[-1]):
        start_indices = np.insert(start_indices, 0, 0)
    min_duration = 0.1 * sr
    valid_segments = [(start, end) for start, end in zip(start_indices, end_indices) if end - start > min_duration]
    merged_segments = []
    prev_start, prev_end = valid_segments[0]
    for start, end in valid_segments[1:]:
        if start - prev_end < merge_gap * sr:
            prev_end = end
        else:
            merged_segments.append((prev_start, prev_end))
            prev_start, prev_end = start, end
    merged_segments.append((prev_start, prev_end))
    padding_samples = int(padding * sr)
    merged_segments = [(max(0, start - padding_samples), min(len(y), end + padding_samples)) for start, end in merged_segments]
    return merged_segments

def compare_with_database(input_index, database):
    distances = [abs(input_index - compute_min_spectrum_index(signal, sr)) for signal in database]
    return np.min(distances)  # Devolver la distancia m√≠nima

# Streamlit UI
st.title("An√°lisis de Audio en Streamlit")

uploaded_file = st.file_uploader("Sube un archivo de audio", type=["mp3", "wav", "ogg", "flac"])

trimmed_signal = np.array([])  # Inicializar trimmed_signal

if uploaded_file is not None:
    st.audio(uploaded_file, format='audio/wav')
    temp_file = "temp_audio.wav"
    with open(temp_file, "wb") as f:
        f.write(uploaded_file.getvalue())

    # Convertir a WAV si es necesario
    converted_file = convert_to_wav(temp_file, "converted_audio.wav")

    # Cargar el audio con librosa
    signal, sr = librosa.load(converted_file, sr=16000, mono=True)

    # Cortar el audio basado en la energ√≠a
    segments = segment_audio_by_energy(signal, sr)
    if len(segments) > 0:
        start, end = segments[0]
        trimmed_signal = signal[start:end]
    else:
        trimmed_signal = signal  # Si no se encuentra ning√∫n segmento, usar la se√±al completa

    # Mostrar forma de onda
    st.subheader("Forma de Onda")
    fig, ax = plt.subplots()
    ax.plot(trimmed_signal)  # Usar trimmed_signal si est√° definido, o signal en su lugar
    ax.set_title("Forma de Onda del Audio")
    ax.set_xlabel("Tiempo")
    ax.set_ylabel("Amplitud")
    st.pyplot(fig)

    # Mostrar espectro
    st.subheader("Espectro de Frecuencia")
    spectrum = compute_spectrum(trimmed_signal, sr)  # Usar trimmed_signal si est√° definido, o signal en su lugar
    fig, ax = plt.subplots()
    ax.plot(spectrum)
    ax.set_title("Espectro de Frecuencia")
    ax.set_xlabel("Frecuencia")
    ax.set_ylabel("Magnitud")
    st.pyplot(fig)

    # Mostrar separabilidad en 2D
    st.subheader("Separabilidad en 2D")
    # Cargar bases de datos para la separabilidad
    with open("correct_database.pkl", "rb") as f:
        correct_database = pickle.load(f)
    with open("error_database.pkl", "rb") as f:
        error_database = pickle.load(f)

    # Combinar bases de datos y el nuevo audio
    X = np.vstack((correct_database, error_database, spectrum.reshape(1, -1)))
    labels = np.array([1] * len(correct_database) + [0] * len(error_database) + [2])  # 2 para el nuevo audio

    # Reducci√≥n de dimensionalidad con t-SNE
    red_ = TSNE(perplexity=20, n_components=2, random_state=123, learning_rate='auto', init='pca')
    X_2D = red_.fit_transform(X)

    # Graficar separabilidad 2D
    fig, ax = plt.subplots(figsize=(10, 8))
    scatter = ax.scatter(X_2D[:, 0], X_2D[:, 1], c=labels, cmap='viridis', s=50)
    plt.colorbar(scatter, label="Etiquetas (1: Correcta, 0: Incorrecta, 2: Nuevo Audio)")
    ax.set_title("Separabilidad de los datos en 2D")
    ax.set_xlabel("Componente 1")
    ax.set_ylabel("Componente 2")

    # Agregar etiqueta de texto para el nuevo audio
    # El nuevo audio es el √∫ltimo punto en X_2D y labels
    ax.text(X_2D[-1, 0], X_2D[-1, 1], uploaded_file.name, fontsize=10, color='red')

    st.pyplot(fig)

    # Calcular el √≠ndice m√≠nimo del espectro
    if len(trimmed_signal) > 0:
        input_min_index = compute_min_spectrum_index(trimmed_signal, sr)

        # Cargar las bases de datos
        with open("correct_database.pkl", "rb") as f:
            correct_database = pickle.load(f)
        with open("error_database.pkl", "rb") as f:
            error_database = pickle.load(f)


        # Comparar con la base de datos de palabras correctas
        distance_correct = compare_with_database(input_min_index, correct_database)

        # Comparar con la base de datos de palabras incorrectas
        distance_error = compare_with_database(input_min_index, error_database)

        # Tomar una decisi√≥n
        if distance_correct < distance_error:
            st.success("‚úÖ El audio pertenece a la palabra correcta.")
        else:
            st.error("‚ùå El audio pertenece a una palabra incorrecta.")
    else:
        st.warning("No se pudo procesar el audio. Aseg√∫rate de que el archivo sea v√°lido.")


#-------------------------------------------------------------------------------------------

```
