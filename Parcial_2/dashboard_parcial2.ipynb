```python
%%writefile 1_ðŸ“ˆ_AnÃ¡lisis_del_Espectro_de_Fourier.py

import streamlit as st

import numpy as np
import matplotlib.pyplot as plt


# ConfiguraciÃ³n de la pÃ¡gina en Streamlit
st.set_page_config(page_title="AnÃ¡lisis del Espectro de Fourier", page_icon="ðŸ“ˆ")

st.markdown("# AnÃ¡lisis del Espectro de Fourier")
st.sidebar.header("ConfiguraciÃ³n de ParÃ¡metros")


# SecciÃ³n TeÃ³rica
st.markdown("## 1. ExpresiÃ³n del Espectro de Fourier")
st.write("El espectro de Fourier para la seÃ±al se obtiene utilizando identidades trigonomÃ©tricas y la serie de Fourier.")

st.markdown("### 1.1. SimplificaciÃ³n de la SeÃ±al")
st.latex(r"x(t) = |A\cos(2\pi F_o t)|^2")
st.write("Usamos la identidad trigonomÃ©trica:")
st.latex(r"\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}")
st.write("Para obtener:")
st.latex(r"x(t) = \frac{A^2}{2} + \frac{A^2}{2} \cos(4\pi F_o t)")

st.markdown("### 1.2. Coeficientes de Fourier")
st.write("La serie de Fourier en forma exponencial es:")
st.latex(r"x(t) = \sum_{n=-\infty}^{\infty} c_n e^{j 2\pi n F_o t}")
st.write("Los coeficientes son:")
st.latex(r"c_0 = \frac{A^2}{2}, \quad c_2 = c_{-2} = \frac{A^2}{4}, \quad c_n = 0 \text{ para otros } n.")
st.write("La representaciÃ³n en forma trigonomÃ©trica es:")
st.latex(r"x(t) = \frac{A^2}{2} + \frac{A^2}{2} \cos(4\pi F_o t)")

st.markdown("### 1.3. Espectro de Fourier")
st.write("El espectro tiene componentes en:")
st.latex(r"f = -2F_o, 0, 2F_o")
st.write("Con magnitudes:")
st.latex(r"\frac{A^2}{4}, \frac{A^2}{2}, \frac{A^2}{4}")
st.write("Y fase 0 en todos los casos.")


#mostrar la seÃ±al
Fo = 60
T = 1/Fo
Fs = 100*Fo # frecuencia de muestreo definida
tv = np.arange(-T/2,T/2,1/Fs) #vector de tiempo generado
Nm = len(tv) #numero de muestras requeridas
A=1 #Constante definida
x = (A*np.cos(2*np.pi*Fo*tv))**2

#graficar seÃ±al original
fig, ax = plt.subplots()
ax.plot(tv, x, 'r', linewidth=4)
ax.grid()
ax.set_xlabel("t [s]", fontsize=14)
ax.set_ylabel("x(t)", fontsize=14)
ax.set_title("GrÃ¡fica de x(t)")

# Mostrar la grÃ¡fica en Streamlit
st.pyplot(fig)

N = 50 #numero armonicos
wo = 2*np.pi/T # frecuencia fundamental
# definir bases
phin = np.zeros((Nm,2*N+1),dtype=np.complex_)#crear matriz para guardar bases
for n in range(-N,N+1,1):
    phin[:,n+N] = np.exp(1j*n*wo*tv)#base de fourier en el intervaloÂ deÂ interÃ©s


#calcular espectro -> cambiar segÃºn la seÃ±al estudiada
cn = np.zeros(2*N+1,dtype=np.complex_)
nv = np.linspace(-N, N, 2*N+1) #vector num armÃ³nicos
Cn = 0*nv
Cn[N] = (A**2)/2 #Nivel DC
Cn[N+2] = (A**2)/4
Cn[N-2]=(A**2)/4

#graficar espectro
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Graficar la parte real
axes[0, 0].stem(nv, np.real(Cn), 'r')
axes[0, 0].set_xlabel(r'$n$', fontsize=14)
axes[0, 0].set_ylabel(r'$Re\{C_n\}$', fontsize=14)
axes[0, 0].grid()
axes[0, 0].set_title("Parte Real")

# Graficar la parte imaginaria
axes[0, 1].stem(nv, np.imag(Cn), 'r')
axes[0, 1].set_xlabel(r'$n$', fontsize=14)
axes[0, 1].set_ylabel(r'$Im\{C_n\}$', fontsize=14)
axes[0, 1].grid()
axes[0, 1].set_title("Parte Imaginaria")

# Graficar la magnitud del espectro
axes[1, 0].stem(nv, abs(Cn), 'r')
axes[1, 0].set_xlabel(r'$n$', fontsize=14)
axes[1, 0].set_ylabel(r'$|C_n|$', fontsize=14)
axes[1, 0].grid()
axes[1, 0].set_title("Magnitud del Espectro")

# Graficar la fase del espectro
phase = np.zeros_like(Cn)
for i in range(len(Cn)):
    if np.imag(Cn[i]) == 0:
        phase[i] = 0  # Fase cero cuando la parte imaginaria es cero
    else:
        phase[i] = np.angle(Cn[i])

axes[1, 1].stem(nv, phase, 'r')
axes[1, 1].set_xlabel(r'$n$', fontsize=14)
axes[1, 1].set_ylabel(r'$\langle C_n$', fontsize=14)
axes[1, 1].grid()
axes[1, 1].set_title("Fase del Espectro")

# Ajustar el diseÃ±o y mostrar en Streamlit
fig.tight_layout()
st.pyplot(fig)



fig, axes = plt.subplots(2, 1, figsize=(8, 8))

# Graficar la magnitud en dB
axes[0].stem(nv, 20 * np.log10(abs(Cn)), 'r')
axes[0].set_xlabel(r'$n$', fontsize=14)
axes[0].set_ylabel(r'$|C_n|$ [dB]', fontsize=14)
axes[0].grid()
axes[0].set_title("Magnitud del Espectro (dB)")

# Graficar la fase ajustada manualmente
phase = np.zeros_like(Cn, dtype=float)
for i in range(len(Cn)):
    if np.imag(Cn[i]) == 0:
        phase[i] = 0  # Fase cero cuando la parte imaginaria es cero
    else:
        phase[i] = np.angle(Cn[i], deg=True)  # Calcular la fase en grados

axes[1].stem(nv, phase, 'r')
axes[1].set_xlabel(r'$n$', fontsize=14)
axes[1].set_ylabel(r'$Fase$ [Â°]', fontsize=14)
axes[1].grid()
axes[1].set_title("Fase del Espectro")

# Ajustar el diseÃ±o y mostrar en Streamlit
fig.tight_layout()
st.pyplot(fig)



# Definir parÃ¡metros iniciales
Px = 0.375  # EstimaciÃ³n de la potencia
N_max = len(Cn) // 2  # MÃ¡ximo valor de Na basado en Cn

# Crear slider en Streamlit
Na = st.slider("Selecciona Na:", min_value=0, max_value=N_max, value=1, step=1)

# Calcular Ã­ndice y error
ind = np.arange(N - Na, N + Na + 1)
er = 1 - np.sum(abs(Cn[ind]) ** 2) / Px

# SeÃ±al reconstruida
xe = phin[:, ind].dot(Cn[ind])  # Producto entre vectores

# Graficar en Streamlit
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(tv, xe, color='b', marker='o', markersize=5, label='$x_e(t)$')  # SeÃ±al estimada
ax.plot(tv, x, color='r', label='x(t)')  # SeÃ±al original
ax.set_title(f'$E_r$ = {100 * er:.2f}%', fontsize=16)
ax.set_xlabel('t [s]')
ax.set_ylabel('x(t)')
ax.grid()
ax.legend()

# Mostrar grÃ¡fica en Streamlit
st.pyplot(fig)



st.markdown("## 4. ConclusiÃ³n")
st.write("""
El anÃ¡lisis del espectro de Fourier muestra que la seÃ±al tiene componentes en $( f = -2F_o, 0, 2F_o )$, con magnitudes $frac{A^2}{4}$, $frac{A^2}{2}$, $frac{A^2}{4}$. La reconstrucciÃ³n de la seÃ±al con una serie finita de tÃ©rminos demuestra que con pocas componentes se puede recuperar la forma original.
""")


```

##TERCER PUNTO

```python
%%writefile 3_âš¡_Total_Harmonic_Distortion_THD.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt


# ConfiguraciÃ³n de la pÃ¡gina
st.set_page_config(page_title="DistorsiÃ³n Total de ArmÃ³nicos (THD)", page_icon="âš¡")

# TÃ­tulo y descripciÃ³n
st.title("DistorsiÃ³n Total de ArmÃ³nicos (THD) y Factor de Potencia")
st.sidebar.header("ConfiguraciÃ³n del Circuito")


st.header("1. DistorsiÃ³n Total de ArmÃ³nicos (THD)")
st.write(
    """
    La **DistorsiÃ³n Total de ArmÃ³nicos (THD)** mide la distorsiÃ³n en una seÃ±al periÃ³dica debido a la presencia de armÃ³nicos.
    Se calcula como:
    """
)
st.latex(r"""
THD = \frac{\sqrt{\sum_{n=2}^{\infty} V_n^2}}{V_1} \times 100\%
""")
st.write(
    "Donde $V_1$ es la magnitud de la frecuencia fundamental y $V_n$ las magnitudes de los armÃ³nicos."
)

st.header("2. CÃ¡lculo del THD desde la FFT")
st.write(
    """
    La Transformada RÃ¡pida de Fourier (FFT) permite calcular el THD descomponiendo la seÃ±al en sus componentes de frecuencia.
    """
)
st.latex(r"""
THD = \frac{\sqrt{V_2^2 + V_3^2 + \dots + V_n^2}}{V_1} \times 100\%
""")

st.header("3. Factor de Potencia y DistorsiÃ³n")
st.write("El factor de potencia (FP) mide la eficiencia en el uso de la energÃ­a:")
st.latex(r"""
FP = \cos(\phi) = \frac{P}{S}
""")
st.write("El **factor de potencia total** se afecta por la distorsiÃ³n de armÃ³nicos, y se puede calcular asÃ­:")
st.latex(r"""
DFP \approx \frac{THD^2}{\sqrt{1 + THD^2}}
""")
st.latex(r"""
FP_{total} = FP_{fundamental} \times \sqrt{1 - DFP^2}
""")

st.header("4. SimulaciÃ³n de THD y Factor de Potencia")
load_type = st.selectbox("Seleccione el tipo de carga:", ["Resistiva", "RC en serie"])
R = st.number_input("Ingrese la resistencia (Î©):", min_value=1.0, value=10.0)
C = None
if load_type == "RC en serie":
    C = st.number_input("Ingrese la capacitancia (F):", min_value=0.000001, value=0.001)

T = 1  # Periodo de la seÃ±al (1 s)
fs = 10000  # Frecuencia de muestreo
t = np.linspace(0, T, fs, endpoint=False)

def rectifier_waveform(t, load_type, R, C=None):
    v_in = np.abs(np.sin(2 * np.pi * t))  # Rectificador de onda completa
    if load_type == "Resistiva":
        i_out = v_in / R
    elif load_type == "RC en serie" and C is not None:
        i_out = np.gradient(v_in, t[1]-t[0]) / C + v_in / R
    else:
        raise ValueError("Carga no vÃ¡lida o parÃ¡metros faltantes")
    return v_in, i_out

v_in, i_out = rectifier_waveform(t, load_type, R, C)

def calculate_thd(signal, harmonics_count=10):
    fft = np.fft.fft(signal) / len(signal)
    fft = np.abs(fft[:len(signal)//2])
    fundamental = fft[1]
    harmonics = fft[2:harmonics_count+2]
    return np.sqrt(np.sum(harmonics**2)) / fundamental * 100

thd = calculate_thd(i_out)
fp_fundamental = 0.9
dfp = (thd / 100)**2 / np.sqrt(1 + (thd / 100)**2)
fp_total = fp_fundamental * np.sqrt(1 - dfp**2)

st.subheader("Resultados de la simulaciÃ³n")
st.write(f"**THD calculado:** {thd:.2f} %")
st.write(f"**Factor de potencia total:** {fp_total:.4f}")

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t, v_in, label="Voltaje de entrada (V_in)")
ax.plot(t, i_out, label="Corriente de salida (I_out)", linestyle="--")
ax.set_title(f"Rectificador de onda completa - Carga: {load_type}")
ax.set_xlabel("Tiempo (s)")
ax.set_ylabel("Amplitud")
ax.legend()
ax.grid()
st.pyplot(fig)

```

```python
%%writefile 2_ðŸ“¡ðŸŽµ_ModulaciÃ³n_en_Amplitud_AM.py


import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import yt_dlp
import soundfile as sf
from io import BytesIO
import tempfile
import os
from scipy.io.wavfile import write
import io

# ConfiguraciÃ³n inicial
st.title("ModulaciÃ³n AM")

# ParÃ¡metros
Fc = 100  # Frecuencia de la portadora (Hz)
Fm = 10   # Frecuencia del mensaje (Hz)
Am = 1    # Amplitud del mensaje

# DefiniciÃ³n del tiempo
ti = 0
tf = 5 / Fm  # NÃºmero de perÃ­odos del mensaje a graficar
Fs = 5 * Fc  # Frecuencia de muestreo
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo

# GeneraciÃ³n de seÃ±ales
m = Am * np.cos(2 * np.pi * Fm * t)  # SeÃ±al del mensaje
Im = 0.8  # Ãndice de modulaciÃ³n
Ac = max(abs(m)) / Im  # Amplitud de la portadora segÃºn el Ã­ndice de modulaciÃ³n
c = Ac * np.sin(2 * np.pi * Fc * t)  # SeÃ±al portadora

y = (1 + m / Ac) * c  # SeÃ±al modulada en AM

# GrÃ¡fico 1: SeÃ±al de la portadora y mensaje
st.subheader("SeÃ±al de la portadora y mensaje")
fig1, ax1 = plt.subplots()
ax1.plot(t, c, label='Portadora')
ax1.plot(t, m, label='Mensaje')
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Amplitud")
ax1.set_title("Portadora y mensaje en el tiempo")
ax1.legend()
st.pyplot(fig1)

# ExplicaciÃ³n del grÃ¡fico 1
st.write("Este grÃ¡fico muestra la seÃ±al de la portadora (onda senoidal de alta frecuencia) y la seÃ±al del mensaje (onda cosenoidal de menor frecuencia). La portadora sirve como la base sobre la cual se modula la informaciÃ³n del mensaje.")

# GrÃ¡fico 2: SeÃ±al modulada en AM
st.subheader("SeÃ±al modulada en AM")
fig2, ax2 = plt.subplots()
ax2.plot(t, y, label='SeÃ±al AM')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Amplitud")
ax2.set_title("SeÃ±al modulada en amplitud")
ax2.legend()
st.pyplot(fig2)

# ExplicaciÃ³n del grÃ¡fico 2
st.write("En este grÃ¡fico se muestra la seÃ±al resultante despuÃ©s de la modulaciÃ³n en amplitud (AM). La envolvente de la seÃ±al modulada sigue la forma de la seÃ±al del mensaje, lo que permite recuperar la informaciÃ³n original en el receptor.")

# Transformada de Fourier
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)

vfre = np.fft.fftfreq(len(c), 1 / Fs)

# GrÃ¡fico 3: Espectro de frecuencia
st.subheader("Espectro de frecuencia de las seÃ±ales")
fig3, ax3 = plt.subplots()
ax3.plot(vfre, abs(Xfy), label='Modulado')
ax3.plot(vfre, abs(Xfm), label='Mensaje')
ax3.plot(vfre, abs(Xfc), label='Portadora')
ax3.set_xlabel("Frecuencia (Hz)")
ax3.set_ylabel("|X(f)|")
ax3.set_title("AnÃ¡lisis espectral de las seÃ±ales")
ax3.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax3.legend()
st.pyplot(fig3)

# ExplicaciÃ³n del grÃ¡fico 3
st.write("Este grÃ¡fico muestra el espectro de frecuencia de las seÃ±ales. Se puede observar la seÃ±al del mensaje en bajas frecuencias, la portadora en su frecuencia correspondiente y la seÃ±al modulada con sus componentes espectrales en torno a la portadora.")

#CanciÃ³n

# ConfiguraciÃ³n de la app
st.title("Descarga y reproducciÃ³n de audio desde YouTube")

# InformaciÃ³n de la canciÃ³n
nombre_cancion = "Queen - Killer Queen"
st.write(f"**CanciÃ³n:** {nombre_cancion}")

# Descarga del audio desde YouTube
def descargar_audio(url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'extractaudio': True,
        'audioformat': 'mp3',
        'outtmpl': 'tema.mp3'
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

def convertir_a_wav():
    os.system("ffmpeg -y -i tema.mp3 output.wav")

# URL del video de YouTube
link = "https://www.youtube.com/watch?v=9K-Jmg93Xj8"
st.write("Descargando audio...")
descargar_audio(link)
st.write("Convirtiendo a WAV...")
convertir_a_wav()

# Cargar el audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

# Mostrar la frecuencia de muestreo
st.write(f"**Frecuencia de muestreo:** {fs} Hz")

# SelecciÃ³n del fragmento de audio (segundos 20 a 25)
inicio, fin = int(fs * 20), int(fs * 25)
segmento_audio = x[inicio:fin]

# Guardar el segmento en un archivo temporal
with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as temp_wav:
    sf.write(temp_wav.name, segmento_audio, fs)
    temp_wav_path = temp_wav.name

# Reproducir el audio en Streamlit
st.audio(temp_wav_path, format="audio/wav")

st.write("Este audio corresponde al segmento entre los segundos 20 y 25 de la canciÃ³n.")

# ConfiguraciÃ³n de la app
st.title("Procesamiento de Audio y ModulaciÃ³n AM")

# Cargar el audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)

# Seleccionar el canal izquierdo y fragmento de audio (segundos 20 a 25)
ti, tf = 20, 25
xpro = x.copy()
xs = xpro[int(ti * fs):int(tf * fs), 0]

# Vector de tiempo
tt = np.arange(ti, tf, 1 / fs)

# GrÃ¡fico 1: SeÃ±al original (canal izquierdo)
st.subheader("SeÃ±al Original del Canal Izquierdo")
fig1, ax1 = plt.subplots()
ax1.plot(tt, xs, label="Canal izquierdo")
ax1.set_xlabel("Tiempo (s)")
ax1.set_ylabel("Amplitud")
ax1.set_title("SeÃ±al de audio entre 20s y 25s")
ax1.legend()
st.pyplot(fig1)

# ExplicaciÃ³n del grÃ¡fico 1
st.write("Este grÃ¡fico muestra la seÃ±al original del canal izquierdo en el intervalo de tiempo entre 20 y 25 segundos.")

# Entrada del usuario para el Ã­ndice de modulaciÃ³n
Im = st.number_input("Ingrese el Ã­ndice de modulaciÃ³n:", min_value=0.1, max_value=2.0, value=0.8, step=0.1)

# ParÃ¡metros de modulaciÃ³n AM
Fc = 1000  # Frecuencia de la portadora (Hz)
Fs = 48000  # Frecuencia de muestreo
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo

# Generar la portadora
Ac = max(abs(xs)) / Im  # Amplitud de la portadora segÃºn Ã­ndice de modulaciÃ³n
c = Ac * np.sin(2 * np.pi * Fc * t)  # SeÃ±al portadora

# GrÃ¡fico 2: SeÃ±al portadora y mensaje
st.subheader("SeÃ±al Portadora y Mensaje")
fig2, ax2 = plt.subplots()
ax2.plot(t, c, label='Portadora')
ax2.plot(t, xs, label='Mensaje')
ax2.set_xlabel("Tiempo (s)")
ax2.set_ylabel("Amplitud")
ax2.set_title("ComparaciÃ³n entre portadora y mensaje")
ax2.legend()
st.pyplot(fig2)

# ExplicaciÃ³n del grÃ¡fico 2
st.write("En este grÃ¡fico se observa la seÃ±al portadora (onda senoidal de alta frecuencia) y la seÃ±al del mensaje original (audio extraÃ­do). La portadora se usa para modular la seÃ±al de audio.")

# SeÃ±al modulada en AM
y = (1 + xs / Ac) * c

# GrÃ¡fico 3: SeÃ±al modulada AM
st.subheader("SeÃ±al Modulada en AM")
fig3, ax3 = plt.subplots()
ax3.plot(t, y, label='SeÃ±al Modulada AM')
ax3.set_xlabel("Tiempo (s)")
ax3.set_ylabel("Amplitud")
ax3.set_title("ModulaciÃ³n AM de la SeÃ±al de Audio")
ax3.legend()
st.pyplot(fig3)

st.write("Este grÃ¡fico muestra la seÃ±al modulada en amplitud (AM). La forma de la seÃ±al refleja cÃ³mo la envolvente sigue la seÃ±al del mensaje, lo que permite transmitir la informaciÃ³n del audio.")

# Transformada de Fourier
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# GrÃ¡fico 4: Espectro de la portadora
st.subheader("Espectro de Frecuencia de la Portadora")
fig4, ax4 = plt.subplots()
ax4.plot(vfre, abs(Xfc), label='Portadora')
ax4.set_xlabel("Frecuencia (Hz)")
ax4.set_ylabel("|X(f)|")
ax4.set_title("Espectro de Frecuencia de la SeÃ±al Portadora")
ax4.legend()
st.pyplot(fig4)
st.write("La seÃ±al portadora es una onda sinusoidal pura de frecuencia Fc = 1000 Hz. Su espectro de frecuencia muestra un Ãºnico pico en esta frecuencia, ya que una onda sinusoidal tiene toda su energÃ­a concentrada en una sola frecuencia")

# GrÃ¡fico 5: Espectro de la SeÃ±al Modulada AM
st.subheader("Espectro de Frecuencia de la SeÃ±al Modulada")
fig5, (ax5a, ax5b) = plt.subplots(2, 1, figsize=(8, 8), constrained_layout=True)  # Aumentando el espacio
ax5a.plot(vfre, abs(Xfy), label='Modulado', color='r')
ax5a.set_xlabel("Frecuencia (Hz)")
ax5a.set_ylabel("|X(f)|")
ax5a.set_title("Espectro de la SeÃ±al Modulada AM")
ax5a.legend()
ax5a.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])

ax5b.plot(vfre, abs(Xfm), label='Mensaje')
ax5b.set_xlabel("Frecuencia (Hz)")
ax5b.set_ylabel("|X(f)|")
ax5b.set_title("Espectro de la SeÃ±al del Mensaje")
ax5b.legend()
ax5b.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])

st.pyplot(fig5)

st.write("El primer grÃ¡fico del espectro muestra la seÃ±al modulada en AM, donde se pueden ver las bandas laterales. El segundo grÃ¡fico muestra el espectro del mensaje original antes de la modulaciÃ³n.")

# FunciÃ³n para convertir el audio en un buffer
def convert_audio(data, rate):
    buffer = io.BytesIO()
    write(buffer, rate, data.astype(np.float32))
    buffer.seek(0)  # Reiniciar el puntero para asegurar la reproducciÃ³n
    return buffer

# Convertir los audios antes de mostrarlos para reducir la demora
buffer_c = convert_audio(c, fs)
buffer_xs = convert_audio(xs, fs)
buffer_y = convert_audio(y, fs)

# Crear pestaÃ±as para seleccionar el audio
tab1, tab2, tab3 = st.tabs(["Portadora", "Mensaje", "Modulada"])

with tab1:
    st.subheader("Audio de la portadora")
    st.audio(buffer_c, format='audio/wav')

with tab2:
    st.subheader("Audio del mensaje original")
    st.audio(buffer_xs, format='audio/wav')

with tab3:
    st.subheader("Audio de la seÃ±al modulada")
    st.audio(buffer_y, format='audio/wav')

#PARTE ESCRITA

# URL de la carpeta de imÃ¡genes en Google Drive
drive_link = "https://drive.google.com/drive/u/1/folders/1jRIc1aQ8pCNZ9DVh_JbURExuuiko-OGv"

# CSS para hacer el botÃ³n flotante
st.markdown(f"""


        ðŸ“¸ Ver Parte escrita

""", unsafe_allow_html=True)
```
