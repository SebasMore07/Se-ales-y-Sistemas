2,1. Semejanzas y diferencias entre la serie de Fourier y las transformadas de Fourier
Serie de Fourier:

1) Exponencial, trigonométrica y compacta:
*   Exponencial: Representa la señal como una suma infinita de términos exponenciales complejos. Es adecuada para análisis matemático.
*   Trigonométrica: Representa la señal como una combinación de senos y cosenos. Es más intuitiva para señales reales.
*   Compacta: Mezcla la forma trigonométrica y exponencial, simplificando las expresiones.

Características:
*   Es válida solo para señales periódicas.
*   El espectro es discreto (con frecuencias armónicas de la fundamental).
*   Sirve para analizar señales periódicas en el tiempo continuo.

2. Transformada de Fourier (FT):

Características:
*   Es una extensión de la serie de Fourier para señales aperiódicas.
* Genera un espectro continuo (amplitudes para todas las frecuencias).
*  Aplica tanto a señales en tiempo continuo como discreto.
*  Convierte una señal del dominio del tiempo al dominio de la frecuencia.

3. Transformada de Fourier en Tiempo Discreto (DTFT):

Características:
*   Aplica a señales discretas en el tiempo, pero produce un espectro continuo en frecuencia.
*   Resulta útil para analizar señales muestreadas.

4. Transformada Discreta de Fourier (DFT):

Características:
*   También aplica a señales discretas, pero produce un espectro discreto (un conjunto finito de frecuencias).
*   Es computacionalmente más práctica que la DTFT para cálculos numéricos, pero introduce aliasing.

2) Espectro continuo vs espectro discreto y señales en tiempo continuo vs discreto

*   Espectro continuo: Se genera cuando las frecuencias de la señal pueden tomar cualquier valor. Es el caso de la FT.
*   Espectro discreto: Aparece cuando las frecuencias son discretizadas (DFT y series de Fourier).
*   Tiempo continuo: Las señales se definen para cualquier instante en el tiempo. (Serie de Fourier y FT en tiempo continuo).
*   Tiempo discreto: Las señales están definidas solo en ciertos instantes. (DTFT, DFT).

3) Fast Fourier Transform (FFT):

¿Qué es?
*   Es un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT).
*   Reduce significativamente el costo computacional del cálculo directo de la DFT.

Costo computacional:
*   DFT directa: 𝑂(𝑁^2), donde 𝑁es el número de muestras.
*   FFT: 𝑂(𝑁log𝑁), lo que la hace mucho más rápida, especialmente para señales grandes.

Funcionamiento:
*   Divide y vencerás: Divide el problema en subproblemas más pequeños y combina los resultados.
*   Se basa en la periodicidad y simetría de las exponenciales complejas.

Utilidad:
*   Procesamiento eficiente de señales digitales.
*   Análisis espectral en tiempo real.
*   Aplicaciones en compresión de datos, imágenes y más.
